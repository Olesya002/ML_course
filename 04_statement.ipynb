```python
import pandas as pd
import sklearn
from sklearn import datasets
import numpy as np
```

# 1. Загрузите выборку Boston

```python
data = datasets.load_boston()
```

```python
# загрузить выборку Boston напрямую мы не можем, воспользуемся альтернативой
data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)
X = pd.DataFrame(np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]]))
y = pd.DataFrame(raw_df.values[1::2, 2])
```

# 2. Приведем признаки в выборке к одному масштабу

```python
from sklearn.preprocessing import scale
```

```python
X_scaled = scale(X)
```

# 3. Переберем разные варианты метрики p по сетке от 1 до 10

```python
p = np.linspace(1, 10, 200)
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import KFold, cross_val_score
```

```python
# создадим генератор разбиений, который перемешивает выборку
kf = KFold(n_splits=5, random_state = 42, shuffle = True)
```

```python
scores = dict()
for i in p:
  knn = KNeighborsRegressor(n_neighbors=5,weights='distance', p = i)
  score = cross_val_score(estimator = knn, X = X_scaled, y = y, cv = kf, scoring = 'neg_mean_squared_error').mean()
  scores[i] = score
print(scores)
```

```python
# индекс максимального элемента
ind = max(scores, key = scores.get)
```

```python
print(ind, scores[ind])
```

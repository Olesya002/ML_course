```python
import pandas as pd
import sklearn
import numpy as np
```

# 1. Загрузим данные цен акций 30 компаний

```python
data = pd.read_csv('close_prices.csv')
data.head()
```

```python
X = data.iloc[:, 1:]
X.head()
```

# 2. На загруженных данных обучим преобразование PCA

```python
from sklearn.decomposition import PCA
```

```python
mc = PCA(n_components=10)
mc.fit(X)
```

```python
1 - mc.explained_variance_ratio_
```

---> хватит 3-х компонент

# 3. Применим построенное преобразование к исходным данным

```python
transformed_X = mc.transform(X)
```

```python
# значения первой компоненты
component_1 = transformed_X[:, 0]
component_1
```

# 3. Загрузим информацию об индексе и посчитаем корреляцию пирсона

```python
index = pd.read_csv('djia_index.csv')
index.head()
```

```python
np.array(index['^DJI'])
```

```python
np.corrcoef(component_1, np.array(index['^DJI']))
```

# 4. Какая компания имеет наибольший вес в первой компоненте

```python
weight_1st_component = mc.components_[0].tolist()
# вклад каждой компании в первую компоненту
```

```python
# ищем индекс компании с максимальным вкладом
X.columns[weight_1st_component.index(max(weight_1st_component))]
```

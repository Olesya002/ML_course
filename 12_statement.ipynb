```python
import pandas as pd
import numpy as np
import sklearn
```

# 1. Загрузим данные по которым будем предсказывать возраст ракушки по физическим измерениям

```python
data = pd.read_csv('abalone.csv')
data.head()
```

# 2. Преобразуем признак Sex в числовой

```python
data['Sex'] = data['Sex'].map(lambda x: 1 if x == 'M' else (-1 if x == 'F' else 0))
```

```python
data.head()
```

# 3. Разделим на признаки и целевую переменную

```python
X, y = data.iloc[:,:-1], data['Rings']
```

```python
X.head()
```

```python
y.head()
```

# 4. Обучим Random Forest с различным числом деревьев и оценим качество модели на кросс-валидации

```python
from sklearn.model_selection import KFold, cross_val_score, cross_val_predict
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
```

```python
kf = KFold(n_splits=5, shuffle=True, random_state=1)
scores = dict()
for n in range(1, 51):
  clf = RandomForestRegressor(n_estimators=n, random_state=1)
  clf.fit(X, y)
  y_pred = cross_val_score(estimator = clf, X = X, y = y, cv = kf, scoring = 'r2').mean()
  scores[n] = y_pred
```

# 5. При каком минимальном кол-ве деревьев лес покажет качество на кросс-валидации выше 0,52

```python
for n in scores:
  if scores[n] >= 0.52:
    print(n)
    break
```

```python
import pandas as pd
import sklearn
```

# 1. Загрузим выборку Wine

```python
data = pd.read_csv('wine.data', names = ['class', 'Alcohol', 'Malic acid', 'Ash', 'Alcalinity of ash', 'Magnesium', 'Total phenols', 'Flavanoids', 'Nonflavanoid phenols', 'Proanthocyanins', 'Color intensity', 'Hue', 'OD280/OD315 of diluted wines', 'Proline'])
data.head()
```

# 2. Извлечем признаки и классы(целевая переменная)

```python
X = data.iloc[:, 1:]
y = data.iloc[:, 0]
```

# 3. Проведем оценку качества методом кросс-валидации по 5 блокам

```python
# создадим генератор разбиений, который перемешивает выборку
```

```python
from sklearn.model_selection import KFold
```

```python
kf = KFold(n_splits=5, random_state = 42, shuffle = True)
```

# 4. Найдите точность классификации на кросс-валидации для метода k ближайших соседей

```python
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
from sklearn.model_selection import cross_val_score
```

```python
# k от 1 до 50
def cross_validation(X_t, y_t):
  scores = []
  for k in range(1,51):
    knn = KNeighborsClassifier(n_neighbors=k)
    score = cross_val_score(estimator = knn, X = X_t, y = y_t, cv = kf, scoring = 'accuracy').mean()
    scores.append(score)
  return scores
```

```python
means = cross_validation(X, y)
accur = pd.DataFrame(means, range(1, 51))
accur[accur.iloc[:,0] == max(means)]
```

# 5. Произведем масштабирование признаков и опять найдем точность классификации

```python
from sklearn.preprocessing import scale
```

```python
X_scaled = scale(X)
```

```python
means_scaled = cross_validation(X_scaled, y)
accur_scaled = pd.DataFrame(means_scaled, range(1, 51))
accur_scaled[accur_scaled.iloc[:,0] == max(means_scaled)]
```

# ---> масштабирование помогло, ведь качество классификации улучшилось
